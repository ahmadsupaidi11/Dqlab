```python
from sklearn.datasets import make_blobs
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
# features menyimpan numpy array 2 dimensi, true label berisikan label cluster dari setiap data
#make_lobs berfungsi untuk membuat sampel data secara acak
features, true_labels = make_blobs(
    n_samples=200, centers=3,
    cluster_std=2,
    random_state=24
)
```

```python
sns.scatterplot(
    x=features[:, 0], y=features[:, 1],
    hue=true_labels)
plt.legend(loc='lower left')
plt.show()
```

```python
#transform filter
standar_scalrt = StandardScaler()
scaled_features = standar_scalrt.fit_transform(features)
```

```python
#mengimplementasikan cluster kmeans dengan membagi data ke dalam 3 cluster
kmeans =KMeans(n_clusters=3, random_state=24)
#menlakukan penghitungan cluster center, kemudian melakukan prediksi
result = kmeans.fit_predict(scaled_features)
```

```python
sns.scatterplot(x=scaled_features[:, 0], y=scaled_features[:, 1], hue=result)
sns.scatterplot(x =
    kmeans.cluster_centers_[:, 0], 
    y = kmeans.cluster_centers_[:, 1],
    s=200, markers='*', color='blue')
plt.legend(loc='lower left')
plt.show()
```
